# TXT 저장 버전

```python
#@title 기본 제목 텍스트
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys 
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import csv


#브라우저열기
browser = webdriver.Chrome('C:\\Users\\ryuji\\OneDrive\\crawler2\\chromedriver.exe')
browser.maximize_window() #화면최대화
browser.implicitly_wait(10) #브라우저 열릴 시간 확보를 위한 10초

#브라우저 주소 갖고오기
browser.get('https://www.instagram.com/')

#주소로 이동시간 확보
sleep(3)


#로그인링크찾기
login_elem = browser.find_element_by_xpath(
	'//*[@id="react-root"]/section/main/article/div[2]/div[2]/p/a')

#로그인 클릭
login_elem.click()

sleep(3)

#로그인정보 입력
browser.find_elements_by_name("username")[0].send_keys("shop_my_baby_f")
browser.find_elements_by_name("password")[0].send_keys("5dnfrkwhr!")

#로그인버튼 찾고
login_button = browser.find_element_by_xpath(
	'//*[@id="react-root"]/section/main/div/article/div/div[1]/div/form/div[4]/button')

#로그인버튼 클릭
ActionChains(browser)\
.move_to_element(login_button)\
.click().perform()
    
browser.find_element_by_xpath("/html/body/div[4]/div/div/div[3]/button[2]").click()

#검색위치찾기
search = browser.find_element_by_css_selector(
	'#react-root > section > nav > div._8MQSO.Cx7Bp > div > div > div.LWmhU._0aCwM > input')

#검색어입력
ActionChains(browser)\
.move_to_element(search).click()\
.send_keys('#소통')\
.perform()

#검색결과클릭
name = browser.find_element_by_xpath(
	'//*[@id="react-root"]/section/nav/div[2]/div/div/div[2]/div[2]/div[2]/div/a[1]/div')

ActionChains(browser)\
.move_to_element(name)\
.click().perform()


sleep(5)

#첫번째 게시글 클릭
post = browser.find_element_by_xpath("//*[@id='react-root']//div[@class='v1Nh3 kIKUG  _bz0w']")
post.click()

#텍스트 파일 쓰기
txt_file = open("#소통.txt",'w')

#첫번째 게시글에서 해시태그 모으기
try :
    for i in range(1,30) :
        tags_xpath = "//div[@class='C4VMK']/span/a" + "[" + str(i) +"]"
        tag =browser.find_element_by_xpath(tags_xpath).text
        if '#' in tag :
            txt_file.write(tag.lstrip('#'))
            txt_file.write(" ")
except :
    another_post = browser.find_element_by_xpath("/html/body/div[4]/div[1]/div/div/a")
    another_post.click()
    
txt_file.write("\n")

#두번째 게시글로 넘어가서 200개의 게시글에서 해시태그 모으기
for i in range(1,200) :
    try :
        for i in range(1,60) :
            tags_xpath =  "//div[@class='C4VMK']/span/a" + "[" + str(i) +"]"
            tag =browser.find_element_by_xpath(tags_xpath).text
            if '#' in tag :
                txt_file.write(tag.lstrip('#'))
                txt_file.write(" ")
                
    except :
        another_post = browser.find_element_by_xpath("/html/body/div[4]/div[1]/div/div/a[2]")
        another_post.click()
        
    txt_file.write("\n")


txt_file.close()
```
